```python
import control as ct
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Parameters
N = 1500 # number of turns
r = 0.017 # radius of the coil [m]
A = r**2 * np.pi # cross-sectional area of the coil [m^2]
mu_0 = 4 * np.pi * 1e-7 # permeability of free space [H/m]
i_0 = 1 # linearization current [A]
m = 0.1 # mass of the magnet [kg]
x_0 = 0.1 # linearization position of the magnet [m]
g = 9.81 # acceleration due to gravity [m/s^2]
R = 28.5 # resistance of the coil [Ohm]

K_P = 20
K_D = 1

G = ct.tf([2*N**2*A*mu_0*i_0], [2*m*x_0**2, 0, 4*m*g*x_0])
C = ct.tf([K_D, K_P], [1])

sys = G*C
print(sys)
ct.root_locus(sys, grid=False)
plt.show()
```
